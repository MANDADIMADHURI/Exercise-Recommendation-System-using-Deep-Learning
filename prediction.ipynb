```python
# In Colab (prediction.ipynb)
!pip install numpy==1.26.4 tensorflow==2.16.1
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Embedding, Flatten, Input
from tensorflow.keras.losses import Huber
from google.colab import files

print("NumPy version:", np.__version__)
print("TensorFlow version:", tf.__version__)

X_train = np.load('/content/drive/MyDrive/Exercise Recommendor/data/X_train.npy')
X_test = np.load('/content/drive/MyDrive/Exercise Recommendor/data/X_test.npy')
y_train_scaled = np.load('/content/drive/MyDrive/Exercise Recommendor/data/y_train_scaled.npy')
y_test_scaled = np.load('/content/drive/MyDrive/Exercise Recommendor/data/y_test_scaled.npy')

vocab_size = 1000
embedding_dim = 16
max_len = 5

input_layer = Input(shape=(max_len * 2,), batch_size=None)  # Explicitly define shape=(10,)
embedding = Embedding(vocab_size, embedding_dim, input_length=max_len * 2)(input_layer)
flatten = Flatten()(embedding)
dense1 = Dense(64, activation='relu')(flatten)
dense2 = Dense(32, activation='relu')(dense1)
output = Dense(1, activation='linear')(dense2)

model = Model(inputs=input_layer, outputs=output)
print("Model input shape:", model.input_shape)  # Should be (None, 10)

model.compile(optimizer='adam', loss=Huber(), metrics=['mean_squared_error', 'mean_absolute_error'])

history = model.fit(X_train, y_train_scaled, validation_data=(X_test, y_test_scaled), epochs=50, batch_size=32)

test_loss, test_mse, test_mae = model.evaluate(X_test, y_test_scaled)
print(f"Test MSE: {test_mse}, Test MAE: {test_mae}")

model.save('/content/drive/MyDrive/Exercise Recommendor/model/exercise_model.h5')
files.download('/content/drive/MyDrive/Exercise Recommendor/model/exercise_model.h5')
```
